```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the real crime dataset
df = pd.read_csv("../data/raw/crime_data.csv")

print(f"Total records: {len(df)}")
df.head()
```

```python
plt.figure(figsize=(8, 5))
df["crime_type"].value_counts().plot(kind="bar", color="steelblue")
plt.title("Crime Type Distribution Across Bangalore")
plt.xlabel("Crime Type")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["hour"], bins=24, kde=False, color="darkorange")
plt.title("Crime Incidents by Hour of Day")
plt.xlabel("Hour (0-23)")
plt.ylabel("Number of Incidents")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df["zone"].value_counts().plot(kind="barh", color="indianred")
plt.title("Crime Incidents by Bangalore Zone")
plt.xlabel("Number of Incidents")
plt.ylabel("Zone")
plt.tight_layout()

plt.show()
```

```python


plt.figure(figsize=(10, 5))
sns.boxplot(x="hour", y="severity", data=df, color="lightseagreen")
plt.title("Crime Severity Distribution by Hour")
plt.xlabel("Hour (0-23)")
plt.ylabel("Severity (1-5)")
plt.tight_layout()
plt.show()
```
